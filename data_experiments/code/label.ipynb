```python
#데이터 분리를 위한 ipynb
import pandas as pd
import pickle

train_base = pd.read_csv("../../../dataset/train/train.csv")
```

```python
train_base
```

```python
train_base["subject_type"] = train_base["subject_entity"].apply(lambda x: eval(x))
```

```python
train_base["subject_type"] = train_base["subject_type"].apply(lambda x: x["type"])
```

```python
train_base
```

```python
train_base["object_type"] = train_base["object_entity"].apply(lambda x: eval(x))
```

```python
train_base["object_type"] = train_base["object_type"].apply(lambda x: x["type"])
```

```python
subject_list = train_base["subject_type"].unique()
```

```python
object_list = train_base["object_type"].unique()
```

```python
object_list
```

```python
for i in subject_list:
    for j in object_list:
        temp = train_base[(train_base['subject_type']==i) & (train_base['object_type']==j)]
        temp = temp[["id", "sentence", "subject_entity", "object_entity", "label", "source"]]
        label_temp = list(temp["label"].unique())
        label_temp.remove('no_relation')
        #num_to_label
        temp_dict = {num : label for num, label in enumerate(label_temp)}
        with open(f'{i}_{j}_num2label.pkl', 'wb') as f:
            pickle.dump(temp_dict, f, protocol=pickle.HIGHEST_PROTOCOL)
        #label_to_num
        temp_dict = {label : num for num, label in enumerate(label_temp)}
        with open(f'{i}_{j}_label2num.pkl', 'wb') as f:
            pickle.dump(temp_dict, f, protocol=pickle.HIGHEST_PROTOCOL)
        temp = train_base[train_base['label'].isin(label_temp)]
        temp = temp[["id", "sentence", "subject_entity", "object_entity", "label", "source"]]
        temp.to_csv(f"{i}_{j}.csv")
```
