```python
from sklearn.model_selection import train_test_split
import pandas as pd

subject_list = ["PER", "ORG"]
object_list = ["DAT", "LOC", "NOH", "ORG", "PER", "POH"]
for i in subject_list:
    for j in object_list:
        so_combine = f"{i}_{j}"
        
        temp = pd.read_csv(f"/opt/ml/dataset/train/{so_combine}.csv")
        x = temp["id"]
        y = temp["label"]
        
        x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, stratify=y)
        temp_train = temp[temp['id'].isin(x_train)][["id", "sentence", "subject_entity", "object_entity", "label", "source"]]
        temp_val = temp[temp['id'].isin(x_val)][["id", "sentence", "subject_entity", "object_entity", "label", "source"]]
        
        temp_train.to_csv(f"{i}_{j}_train.csv", index=False)
        temp_val.to_csv(f"{i}_{j}_val.csv", index=False)
```
